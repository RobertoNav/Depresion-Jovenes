## ADEP

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("../data/raw/adepcuestionarioinicialanual2022_ok.csv")
df2 = pd.read_csv("../data/raw/adepcuestionarioinicialanual2023_ok.csv")
df3 = pd.read_csv("../data/raw/adepcuestionarioinicialanual2024_ok.csv")
df4 = pd.read_csv("../data/raw/adepcuestionarioinicialanual2025_ok.csv")

df = pd.concat([df1,df2],axis=0, ignore_index=True)
df = pd.concat([df,df3],axis=0, ignore_index=True)
df = pd.concat([df,df4],axis=0, ignore_index=True)
```

```python
df.info()
```

```python
df.to_csv('../data/interim/base_datos_adep_2022-2025.csv', index=False, encoding='utf-8')
```
